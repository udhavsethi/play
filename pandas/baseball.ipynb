```python
home = pd.DataFrame(df["HomeStartingPitcher"])
home.rename(columns={"HomeStartingPitcher": "Pitcher"}, inplace=True)
home["win"] = np.where((df.FinalScoreHome > df.FinalScoreAway), 1, 0)
```

```python
away = pd.DataFrame(df["AwayStartingPitcher"])
away.rename(columns={"AwayStartingPitcher": "Pitcher"}, inplace=True)
away["win"] = np.where((df.FinalScoreHome < df.FinalScoreAway), 1, 0)
```

```python
home, away
```

```python
pitchers = pd.concat([home, away])
pitchers['num'] = [1] * len(pitchers)
```

```python
pitchers.head()
```

```python
pitchers = pitchers.groupby('Pitcher').sum()
```

```python
pitchers['win_percent'] = 100 * (pitchers['win'] / pitchers['num'])
```

```python
pitchers
```

```python
pitchers.sort_values(by=['win_percent', 'win'], ascending=False)
```
